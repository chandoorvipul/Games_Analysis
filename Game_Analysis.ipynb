```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("PS4_GamesSales.csv", encoding = 'latin1')
data.columns = map(str.lower, data.columns)
data = data.dropna()
data
```

```python
data.describe()
```

```python
size = data['genre'].value_counts()
print(size)
```

```python
print(data.isnull().sum())
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='genre', data=data)
plt.xticks(rotation=90)
plt.xlabel('Genre', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.title('Count of Genre according to category', fontsize=15, color='r')
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='year', data=data)
plt.xticks(rotation=90)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.title('Number of games released in the year', fontsize=15, color='r')
```
